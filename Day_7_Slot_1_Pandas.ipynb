```python
import pandas as pd
url = 'https://raw.githubusercontent.com/ramar92/TNSDC-Skill4Future/main/co2emission_data1.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.tail()
```

```python
# Selecting, Indexing & Slicing

# Select Columns: df['column_name']
df['Total Power (MW)']
```

```python
df.columns
```

```python
# Select Multiple Columns: df[['col1', 'col2']]
df[['Day','Total Power (MW)', 'Total Emission (tons)']].head()
```

```python
# Select Rows by Index: df.iloc[0] (first row), df.iloc[5:10] (rows 5-9)
df.iloc[1]
```

```python
# iloc[rows, columns]

df.iloc[5:10, 0:3]
```

```python
# Select by Label: df.loc[5] (row with index 5), df.loc[5, 'col_name']

# loc[row, columns]

df.loc[5:10, 'Day':'Total Power (MW)']
```

```python
# Add New Column: df['New_Column'] = df['CO2'] * 1.1

df['Efficency (%)'] = df['Total Emission (tons)'] / df['Total Power (MW)']
df.head()

```

```python
# Drop Column: df.drop('Column_Name', axis=1, inplace=True)
df.drop('Efficency (%)', axis=1, inplace=True)
df.head()
```

```python
# Drop Row: df.drop(5, axis=0, inplace=True)

df.drop(5, axis=0, inplace=True)
df.head()
```

```python
# Handling Missing Data

1. Check Missing Values: df.isnull().sum()
2. Fill Missing Values: df.fillna(value)
3. Drop Missing Values: df.dropna()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```
