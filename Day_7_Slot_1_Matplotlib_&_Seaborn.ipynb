```python
url = 'https://raw.githubusercontent.com/ramar92/TNSDC-Skill4Future/main/co2emission_data1.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.fillna(0)
```

```python
# 1. Line Plot
plt.plot(df.index, df['Wind Emission (tons)'], marker='o', linestyle='-', color='g', label='Wind Emission')
plt.xlabel('Index')
plt.ylabel('Wind Emission (tons)')
plt.title('Wind Emission Over Time')
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 30]

plt.bar(categories, values, color='g')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x=df.index, y = df['Wind Emission (tons)'], marker='o', color='g', label='Wind Emission')
plt.xlabel('Index')
plt.ylabel('Wind Emission (tons)')
plt.title('Wind Emission Over Time')
plt.show()
```

```python
df['Wind Emission (tons)'] = df['Wind Emission (tons)'].fillna(0)
```

```python
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=df.index, y=df['Wind Emission (tons)'],color='skyblue')
plt.xlabel('Index')
plt.ylabel('Wind Emission (tons)')
plt.title('Wind Emission Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram (Data Distribution)
df['Wind Emission (tons)'] = df['Wind Emission (tons)'].fillna(0)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Coal Power (MW)'], bins=10, color='red', edgecolor='white')
plt.xlabel('Coal Power (MW)')
plt.ylabel('Frequency')
plt.title('Distribution of Wind Emission')
plt.show()
```

```python
sns.histplot(df['Coal Power (MW)'],bins=20, kde=True, color='red', alpha=0.5, label='Seaborn')
plt.xlabel('Coal Power (MW)')
plt.ylabel('Frequency')
plt.title('Distribution of Wind Emission')
plt.show()
```

```python
## Scatter Plot (Relationships)
# Solar Emission (tons)	Wind Emission (tons)

df['Wind Emission (tons)'] = df['Wind Emission (tons)'].fillna(0)
df['Solar Emission (tons)'] = df['Solar Emission (tons)'].fillna(0)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Wind Emission (tons)'], df['Solar Emission (tons)'], color='blue', alpha=0.6, label='Matplotlib')
plt.xlabel('Wind Emission (tons)')
plt.ylabel('Solar Emission (tons)')
plt.title('Relationship between Wind and Solar Emission')
plt.legend()
plt.show()
```

```python
sns.scatterplot(x=df['Wind Emission (tons)'], y=df['Solar Emission (tons)'], color='red', alpha=0.7, label='Seaborn')
```

```python
# Pie Chart
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct="%1.1f%%",colors=["blue", "green", "red", "orange"])
plt.title("Pie Chart Example")
plt.show()
```

```python
#box-plot
sns.boxplot(x=df['Wind Emission (tons)'])
```
